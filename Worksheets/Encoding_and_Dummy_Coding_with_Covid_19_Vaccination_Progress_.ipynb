# **PART 3: ENCODING DATA**

In this section, we will apply One-Hot Encoding approach to creat a new column (sometimes called a dummy variable) with binary encoding (0 or 1) to demote whether a perticular row belongs to this category. 

I am going to write a function to encode "daily vaccinations per million", where values greater than or equal to median = 1 and values less than median = 0 

**3.1 : Encode the daily vaccinations per million**

````python
# to calculate the median for daily_vaccinations_per_million
median = dropna_df["daily_vaccinations_per_million"].median() 
median 

# to encode daily vaccinations per million, where values greater than or equal to median = 1 and values less than median = 0
def encode_daily(df, median):
  if df["daily_vaccinations_per_million"] >= median:
    return 1 
  else: 
    return 0

# to creat a new cloumn "encoded" to hold the encoded values
dropna_df["encoded"] = dropna_df.apply(encode_daily, axis = 1, median = median)
dropna_df # to print the table 
dropna_df["encoded"].describe() # to check the results
````
